```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import seaborn as sns
sns.set(color_codes=True)
sns.set_style("white")

np.random.seed(1337)

#generate flip
def flip(num=1):
    flips =[]
    
    for i in range(num):
        
        num = np.random.uniform(low=0.0 , high=1.0)
        #Assume Head is  "num" greater then 75
        #Less then num <75 is Tail
        if num>0.75:
            flips.append('H')
        else:
            flips.append("T")
    return flips


flips = flip(10)
print(flips)
#every run times we will get differnt prob of head and tail 

values , counts  = np.unique(flips , return_counts=True)
print("Values : ", values , "Counts : ",counts) 

```

# Proability of flips


```python
from  collections import Counter , defaultdict

def get_freqs(flips):
    '''get_freqs function do the calculating of chances and save 
    into the dictionary in the from of head/tails as the key and valeus 
    are counter of that'''
    keys = Counter(flips).keys()
    vals = Counter(flips).values()
    
    print('Key : ', keys)
    print("Vals : ", vals)
    
#     return dict(zip(keys,vals))

    return defaultdict(int,dict(zip(keys, vals)))
```

```python
freqs = get_freqs(flips)
print(freqs)
help(get_freqs)
```

```python
prob_h = freqs['H']/len(flips)
print(prob_h)
```

## Experiment: Prob calculated based on 1 flip upto N flips 

```python
maximum_flips = 1000
prob = []
for num_flips in range(1,maximum_flips):
    flips = flip(num_flips)
    freqs = get_freqs(flips)
    prob_h = freqs['H']/len(flips)
    
    prob.append(prob_h)
print(prob)
```

```python
print(freqs)
```

```python
plt.plot(prob)
plt.ylabel("Estimated Proability ")
plt.xlabel("Number of flips in Experiment")
plt.title("Calculating Probabilty over Numbers of flips")
sns.despine(offset=10 , trim=True) #move the Axes
plt.show()
```

```python
#frist Two element
plt.plot(prob[:10])
plt.ylabel("Estimated Proability ")
plt.xlabel("Number of flips in Experiment")
plt.title("Calculating Probabilty over Numbers of flips")
sns.despine(offset=10 , trim=True) #move the Axes
plt.show()
```

```python
plt.plot(prob[maximum_flips-10:])
plt.ylabel('Estimated Probability')
plt.xlabel('Number of Flips in Experiment');
plt.title("Calculating Probability over Number of Flips")
sns.despine(offset=10, trim=True);  # move axes away 
plt.show()
```
