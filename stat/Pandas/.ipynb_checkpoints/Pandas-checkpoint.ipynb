```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Data/nhanes_2015_2016.csv" ,index_col=0 ,parse_dates=True )
data
```

```python
data.head(10)
```

```python
extract_data=data.iloc[1:10 ,1:10]
extract_data

```

```python
extract_data.plot()
```

```python
extract_data['SMQ020'].plot()
```

```python
extract_data.plot.scatter(x="SMQ020" , y="ALQ130" , alpha=0.6)
```

```python
extract_data.plot.scatter(x="SMQ020" , y="ALQ130" , alpha=0.5)
```

```python
extract_data.plot.bar()
extract_data.plot.area()
extract_data.plot.barh()
extract_data.plot.line()
extract_data.plot.box()
```

```python
extract_data.plot.area(figsize=(15,10) )
```

```python
extract_data.plot.area(figsize=(15,10),subplots=True )
```
